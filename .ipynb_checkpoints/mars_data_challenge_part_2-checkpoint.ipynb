```python
# Import dependencies for WebScraping, Pandas and Matplotlib

from splinter import Browser
from bs4 import BeautifulSoup as soup
from webdriver_manager.chrome import ChromeDriverManager

import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Set up Splinter
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
# Visit Mars Temperature Data
url = 'https://data-class-mars-challenge.s3.amazonaws.com/Mars/index.html'
browser.visit(url)

# Parse the HTML
html = browser.html
mars_soup = soup(html, 'html.parser')
```

```python
# Scrape table using Pandas

df = pd.read_html(url)
print(df)
```

```python
# Convert to Pandas dataframe

df_mars = df[0]
df_mars
```

```python
# Examine the data types of the DataFrame columns

df_mars.info()
```

```python
# Change the date data type

df_mars['terrestrial_date'] = pd.to_datetime(df_mars['terrestrial_date'])
```

```python
# Re-examine the data types of the DataFrame columns and verify the updates

df_mars.info()
```

```python
# How many months exist on Mars?

no_months = len(pd.unique(df_mars['month']))
print(no_months)
```

```python
# How many Martian (and not Earth) days worth of data exist in the scraped dataset?

days = df_mars['sol'].count()
print(days)
```

```python
# What are the coldest and the warmest months on Mars (at the location of Curiosity)?

df_mars.groupby('month').min_temp.mean().plot(
    ylabel = 'temperature',
    grid = True,
    title = 'Mars Average Temperature',
    figsize = (15,5))

plt.show()
```

```python
# Which months have the lowest and the highest atmospheric pressure on Mars?

df_mars.groupby('month').pressure.mean().plot.bar(
    ylabel = 'pressure',
    title = 'Mars Average Pressure',
    figsize = (15,5))

plt.xticks(rotation = 360)

plt.show()
```

```python
# About how many terrestrial (Earth) days exist in a Martian year?  687
#That is, in the time that Mars circles the Sun once, how many days elapse on Earth?
#Visually estimate the result by plotting the daily minimum temperature.

df_mars.min_temp.plot(figsize=(15,5),grid=True)


plt.legend(['687 Earth Days'], prop={'size': 12})

plt.show()
```

```python
# Export dataframe to CSV

df_mars.to_csv('Mars_Data.csv')
```
